**策略生命周期** 
1. **conception： \_\_init\_\_()**  
    实例化过程中调用， indicators 以及其他需要的特征都在这里创建  
2. **birth： start**  
    cerebro 告诉 strategy 需要开始计算了，默认是个空方法  
3. **childhood： prenext**  
    在 conception 中声明的指标， 会在回测日期中加一些限制（how long the strategy needs to mature），被称作  **minimum period**  
    比如在 \_\_init\_\_() 中计算5日均线指标，minimum period 则为5d，默认是空函数  
4. **adulthood：next**  
    minimum period 结束后，策略开始执行，这时候，会有方法 **nextstart** 被执行一次（exactly once），用来转换 prenext 到 next，默认的 nextstart 就只是简单的调用next  
5. **reproduction: None**  
    如果设置了不同的参数，可能会运行多次  
6. **death：stop**  
    默认是个空方法
